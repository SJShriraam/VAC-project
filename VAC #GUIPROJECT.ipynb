```python
import tkinter as tk

class FlightBookingGUI:
    def __init__(self, master):
        self.master = master
        master.title("Flight Booking System")

        # Create labels and entry widgets for flight booking form
        tk.Label(master, text="From: ").grid(row=0, column=0)
        self.from_entry = tk.Entry(master)
        self.from_entry.grid(row=0, column=1)

        tk.Label(master, text="To: ").grid(row=1, column=0)
        self.to_entry = tk.Entry(master)
        self.to_entry.grid(row=1, column=1)

        tk.Label(master, text="Departure Date: ").grid(row=2, column=0)
        self.departure_date_entry = tk.Entry(master)
        self.departure_date_entry.grid(row=2, column=1)

        tk.Label(master, text="Return Date: ").grid(row=3, column=0)
        self.return_date_entry = tk.Entry(master)
        self.return_date_entry.grid(row=3, column=1)

        tk.Label(master, text="Number of Passengers: ").grid(row=4, column=0)
        self.num_passengers_entry = tk.Entry(master)
        self.num_passengers_entry.grid(row=4, column=1)

        # Create button for flight search
        self.search_button = tk.Button(master, text="Search Flights", command=self.search_flights)
        self.search_button.grid(row=5, column=0, columnspan=2)

        # Create label for flight search results
        self.results_label = tk.Label(master, text="")
        self.results_label.grid(row=6, column=0, columnspan=2)

    # Function to search for flights and display results
    def search_flights(self):
        # Get form values
        from_location = self.from_entry.get()
        to_location = self.to_entry.get()
        departure_date = self.departure_date_entry.get()
        return_date = self.return_date_entry.get()
        num_passengers = self.num_passengers_entry.get()

        # Perform flight search and get results
        # This is where you would call an API or perform a database query to get flight information
        # For the purposes of this example, we will just display a message with the search criteria
        search_results = f"Flights from {from_location} to {to_location} on {departure_date}"
        if return_date:
            search_results += f" and returning on {return_date}"
        search_results += f" for {num_passengers} passengers"
        self.results_label.config(text=search_results)

if __name__ == "__main__":
    root = tk.Tk()
    gui = FlightBookingGUI(root)
    root.mainloop()
```
